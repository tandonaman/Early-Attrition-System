```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\apatr\Desktop\SE Project\final_job_role_match.csv")
initial_data = pd.DataFrame(data)
initial_data.head()
```

```python
copied_data = initial_data.copy()
```

```python
copied_data_2 = copied_data.drop(columns=["Unnamed: 2", "Unnamed: 3"])
copied_data_2.head()
```

```python
copied_data_2['Job_Roles'].value_counts()
```

```python
filter_data = copied_data_2[copied_data_2['Job_Roles'] !='NONE']
filter_data.shape
```

```python
filter_list = filter_data['Job_Roles']
list1 = []
for x in filter_list:
    list1.append(x.lower())
    
filter_data['JOB_Role'] = list1
filter_data.shape
```

```python
filter_data1 = filter_data.drop(['Job_Roles'], axis=1)
filter_data1.head()
```

```python
#filter_data1['JOB_Role'].value_counts()
records = filter_data1.to_records(index=False)
result = list(records)
#print(result)
result[0]
```

```python
designation = []
for x in result:
    designation.append(x[1])
    
len(designation)
```

```python
# tHIS IS TO REMOVE ALL THE REDUNDANT DATA. Note here designation holds only job roles.
res = []
for i in designation: 
    if i not in res: 
        res.append(i) 
```

```python
len(res)
# Unique designations in all that data.
```

```python
listing = []
for y in res:
    tempo = []
    for t in result:
        if t[1] == y:
            tempo = tempo + t[0].strip('][').split(', ')
    listing.append(tempo)
    
len(listing)
```

```python
listing[0]
```

```python
# Now removing redundant skill sets for each and every job role.
res_temp = []
final_listing = []
for i in listing:
    for val in i:
        if val not in res_temp: 
            res_temp.append(i) 
    final_listing.append(res_temp)
    
len(final_listing)
```

```python
# Finally I make the dataframe. Note designation holds job roles and final_listing holds skillsets.
dic = {"Job_Roles": res[0], "Skillsets": final_listing[0]}
df = pd.DataFrame(dic) 
df.head()
```
